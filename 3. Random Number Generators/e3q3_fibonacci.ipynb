#### Fibonacci Generator


```python
import matplotlib.pyplot as plt
```

```python
def rng(N0, a, b, M, nsamples):

    U = []
    N = N0

    for i,e in enumerate(range(nsamples)):
        N = (a*N + b) % M
        U.append(N)
    return U
```

```python
def frng(initial, u, v, M, nsamples):

    U = initial

    for i,e in enumerate(range(nsamples)):
        N = (U[i - v] - U[i - u]) % M
        U.append(N)
    U = [i/M for i in U]
    return U
```

```python
def scatter_square(a, b, u, v, M, N, nsamples=None):
    if nsamples is None:
        nsamples = M

    initial_lst = rng(N, a, b, M, v)
    frng_lst = frng(initial_lst, u, v, M, nsamples)
    x = frng_lst[:-1]
    y = frng_lst[1:]

    plt.figure(figsize=(8, 5))
    plt.xlabel('U\u2081')
    plt.ylabel('U\u2099\u208B\u2081')
    plt.plot(x, y, 'rx')
    plt.title(f'Unit Square Scatter of Fibonacci (nsamples={nsamples}, a={a}, b={b}, M={M})')
    plt.show()

scatter_square(1597, 51749, 5, 17, 644980, 24234, 2000)
```
