#### Box and Muller


```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def box_muller(N):
    Z = []
    for i in range(N):
        x1 = random.uniform(0, 1)
        x2 = random.uniform(0, 1)
        squ_root = np.sqrt(-2 * np.log(x1))
        cos2 = np.cos(2 * np.pi * x2)
        sin2 = np.sin(2 * np.pi * x2)
        y1 = squ_root * cos2
        y2 = squ_root * sin2
        Z.append(y1)
        Z.append(y2)
    # print(Z)
    return Z
```

```python
a = 5
plt.figure(figsize=(8, 5))
plt.hist(box_muller(200000), bins=50, range = (-a, a), edgecolor='black', alpha=0.7, density=True, label='Box Muller')
x = np.linspace(-a, a, 1000)
plt.plot(x, norm.pdf(x), 'r-', label='Normal Distribution')
plt.title(f'Histogram of Box Muller')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(False)
plt.show()
```
