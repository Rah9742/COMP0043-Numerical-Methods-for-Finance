```python
import matplotlib.pyplot as plt
import numpy as np

e = 1

print("{:0.22f}".format(e))
```

```python
e = 1

for i in range(100):
    e /= 2
    if e + 1 == 1:
        print(i)
        print("{:0.22f}".format(e*2))
        break
```

```python
n = 100
x_points = np.zeros(n, dtype=float)
y_points = np.zeros(n, dtype=float)

for e in range(1, n + 1):
    index = e - 1
    for i in range(n):
        e /= 2
        if e + 1 == 1:
            x_points[index] = index
            y_points[index] = i
            break

plt.scatter(x_points, y_points)
plt.show()
```

```python
plt.scatter(x_points, 0.5**x_points)
plt.show()
```
